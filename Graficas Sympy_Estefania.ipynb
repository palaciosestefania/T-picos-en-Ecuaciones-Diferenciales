## Curvas paramétrizadas

### Importando plot_parametric desde sympy

```python
#"Una curva parametrizada es una representación matemática de una curva en un espacio n-dimensional. Las ecuaciones paramétricas de la curva se definen en un intervalo y utilizan un parámetro t para describir la posición de la partícula en cualquier tiempo". (proyectodescartes.org)
```

```python
from sympy import plot_parametric #Para graficar curvas paramétricas en R2 y R3
```

```python
s=symbols('s') #definición de variable
#Cuando trabajamos con funciones parametricas trabajamos a su vez con la longitud de arco definida como la integral desde un t0 hasta t, de la derivada de la función parametrica respecto al tiempo.
```

```python
plot_parametric(cos(s),sin(s+1)) #Grafico de xy, con x correspondiente a la función cos(s), y y a la función sin(s+1)
```

```python
plot_parametric((cos(s),sin(s+1)),(s,s**2),(s,-4,4)) #Grafico de dos funciones parametricas con la variable s en el intervalo [-4,4]
```

### Importando plot3d_parametric_surface desde sympy.plotting

```python
from sympy.plotting import plot3d_parametric_surface #para graficar superficies parametricas en R3: f(x,y)
```

```python
u, v =symbols('u v')  #declaración de variables
```

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v) 
#superficie ondulatoria (dada por la funcion cos y sin), siendo u-v la pendiente en el eje z
```

```python
plot3d_parametric_surface(cos(u-v),sin(u-v),v) 
#(cos(u−v),sin(u−v)) describe un círculo unitario, v describe cómo aumenta la superficie con respecto al eje z
#"la superficie estará estratificada en niveles según el valor de v"
```

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
#toro o esfera
#CONSULTADO EN CHAT GPT:
#el primer argumento proyecta un círculo de radio cos⁡(𝑣) en el eje 𝑥, con ángulo u girando alrededor de ese círculo
#el segundo argumento escribe la misma proyección en el eje y creando un círculo en el plano xy
#el tercer argumento Controla la rotación del círculo en el eje z, lo que genera el giro que forma el toro.
```

```python
#ejemplo de código de chat gpt para simulación de la Superficie de Riemann
r, theta = symbols('r theta')

# Coordenadas paramétricas para la superficie de Riemann de w = sqrt(z)
x = r * cos(theta)           # Parte real de z
y = r * sin(theta)           # Parte imaginaria de z
z = sqrt(r) * cos(theta / 2) # Parte real de w = sqrt(z)

# Graficar la superficie de Riemann de sqrt(z)
plot3d_parametric_surface(x, y, z, (r, 0, 20), (theta, 0, 8*pi))
```

#### Ejercicio I. Una partícula se mueve a lo largo del eje x de modo que su velocidad instantánea está dada como una función del tiempo t por v=12-3t^2. Al tiempo t=1 está localizada en x=-5.

```python
import sympy as sp #métodos numéricos
import numpy as np
import scipy as pp #calculos simbólicos
```

```python
from sympy import *
```

##### a) Establezca un problema de valor incial que describa el movimiento.

```python
t=sp.symbols('t') #variable independiente
x, y = sp.symbols('x y',cls=Function) #variables dependientes
```

```python
ode=sp.Eq(x(t).diff(t),12-3*t*t) #expresión que describe el movimiento
```

```python
ode #impresión de la expresión
```

##### b) Resuelva el problema en a).

```python
sp.classify_ode(ode) #clasificación de la expresión dada
```

```python
sol=sp.dsolve(ode) #resolver la ecuación diferencial
```

```python
sol #impresión de la solución GENERAL
```

```python
#Condiciones iniciales x(1)=-5
C1= sp.symbols('C1')
constant = sol.rhs.subs(t, 1)
C = sp.solve(sp.Eq(constant, -5), C1) 
```

```python
C
```

```python
#Solución particular
xsoln=sol.subs(C1, C[0])
xsoln
```

##### c) Determine en dónde estará la partícula en los tiempos t=2 y t=3.

```python
xsol1 = xsoln.subs(t, 2) #sustituye el valor de t=2 en la expresión
xsol1
```

```python
xsol2 = xsoln.subs(t, 3) #sustituye el valor de t=3 en la expresión
xsol2
```

#### Ejercicio II. Una partícula se mueve a lo largo del eje x de modo que su aceleración instantánea está dada como una función del tiempo t por a=10-12t^2. En los tiempos t=2 y t=3 la partícula está localizada en x=0 y x=40, respectivamente.

```python
from sympy.plotting import plot #plotting grafica y manipula graficos, mientras que plot (añadido desde plotting) gráfica y manipula funciones en R2
```

##### a) Establezca la ecuación diferencial y condiciones asociadas que describen el movimiento, ¿El problema es de valor inicial o de frontera?

```python
t=sp.symbols('t') #variable independiente
v, y = sp.symbols('v y',cls=Function) #variables dependientes
```

```python
ode2=sp.Eq(v(t).diff(t),10-12*t**2) #expresión que describe el movimiento
ode2
```

```python
#El problema es de forntera, ya que tenemos valores en puntos distintos como condiciones.
```

##### b) Solucione el problema en a).

```python
sol3=sp.dsolve(ode2) #resolver la ecuación diferencial
sol3 #SOLUCIÓN GENERAL PARA LA VELOCIDAD
```

```python
sp.classify_ode(ode2)
```

```python
#Las condiciones iniciales las tenemos con repecto a x(2)=0 y x(3)=40, por lo que debemos integrar sol3 para trabajr con el movimiento.
```

```python
C1, C2 = sp.symbols('C1 C2') #constantes
v=C1 - 4*t**3 + 10*t
```

```python
x = sp.integrate(v, t)+C2 #integración de v(t)
x #IMPRESIÓN DE LA SOLUCIÓN GENERAL PARA EL MOVIMIENTO
```

```python
x1 = sp.Eq(x.subs(t, 2), 0)  # x(2) = 0
x2 = sp.Eq(x.subs(t, 3), 40)  # x(3) = 40
solp = sp.solve([x1, x2], [C1, C2]) # Resolver el sistema de ecuaciones para C1 y C2
solp
```

```python
solpart= x.subs(solp) #Sustitución de las constantes
```

```python
solpart #impresión de la SOLUCIÓN PARTICULAR
```

##### c) Determine la posición de la partícula en t=1.

```python
x1 = solpart.subs(t, 1) #sustitución de  t=1 en la solución particular DE MOVIMIENTO
x1
```
