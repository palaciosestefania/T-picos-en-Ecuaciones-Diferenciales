```python

import sympy as sp #métodos numéricos
import numpy as np #vectores
import scipy as pp #calculos simbólicos
import matplotlib as mpl
import matplotlib.pyplot as plt

#from sympy import init_printing
#sp.init_printing()
```

```python
x=sp.symbols('x')
y=sp.Function('y')
```

```python
ode=sp.Eq(y(x)*sp.Derivative(y(x),x)+2*x,0)
ode
```

```python
sp.classify_ode(ode)
```

```python
sol=sp.dsolve(ode,y(x))
sol
```

```python
def f(x,c):
    return np.sqrt(c-2*x**2)
```

```python
f(1,4)
```

```python
def g(x,c):
    return np.sqrt(c*x)
```

```python
#grafica solo una
xx=np.linspace(-40,40,4000)
```

```python
for c in ([200,300,400,800,900,1000,1100,1500]):
    plt.plot(xx,f(xx,c),color="cyan")
    plt.plot(xx,-f(xx,c),color="cyan")
    plt.plot(xx,g(xx,c),color="pink")
    plt.plot(xx,-g(xx,c),color="pink")
    plt.plot(-xx,-g(xx,c),color="pink")
    plt.plot(-xx,g(xx,c),color="pink")
plt.ylim(-50,50)
```
